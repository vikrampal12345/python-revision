Take three numbers and print the smallest...

```python
a=int(input("Enter the first number: "))
b=int(input("Enter the second number: "))
c=int(input("Enter the third number: "))

if a<b or a<c:
    print("first number is smaller: ",a)
elif b<a or b<c:
    print("Second number is smallest: ",b)
else:
    print("Third number is smallest: ",c)        
```

using function

```python
def comparison(a,b,c):
    if a<b and a<c:       # and is used because we want to all the conditions must be true
        return f"{a} is smallest number"
    if b<c and b<a:
        return f"{b} is smallest number"
    else:
        return f"{c} is smallest number"
    
a=int(input("Enter the first number: "))
b=int(input("Enter the second number: "))
c=int(input("Enter the third number: "))    

print(comparison(a,b,c))
```

using class and mastery level approach

```python
class comparison:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c

    def show(self):
        smallest=min(self.a,self.b,self.c)
        return f"{smallest} is the smallest number"       
    
a=int(input("Enter the first number: "))    
b=int(input("Enter the second number: "))    
c=int(input("Enter the third number: "))    

obj=comparison(a,b,c)
print(obj.show())
```

using pro approach

```python
class Comparison:
    def __init__(self, *nums):
        self.nums = nums

    def smallest(self):
        return min(self.nums)

a=list(map(int, (input("Enter the number: ").split())))      # we write it in single line
obj=Comparison(*a)    
print(obj.smallest())
```

using function

```python
def factorial(a):
    total=1
    for i in range(1,a+1):
        total *=i
    return total
a=int(input("Enter the number: "))
print(factorial(a))    
```
